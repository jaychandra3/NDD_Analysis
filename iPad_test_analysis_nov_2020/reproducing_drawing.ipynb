```
# import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import matplotlib.animation as animation
from matplotlib.collections import LineCollection
```

```
# reproduce a drawing of the PaHaW spiral 
data = pd.read_csv('/home/jovyan/work/PaHaW_spirals/00082__1_1.csv')
xcoor = []
ycoor = []

for i in data.iloc[:500, 0]: 
    xcoor.append(i)
for i in data.iloc[:500, 1]: 
    ycoor.append(i)
fig = plt.figure(num = 1, clear = True)
ax = fig.add_subplot(1, 1, 1)

ax.scatter(xcoor, ycoor)
```

```
# frame interval - interval between frames defaulted to be 20ms
def draw(data_csv, gif_path, frame_interval = 20): 
    data = pd.read_csv(data_csv)

    fig = plt.figure()
    ax = plt.axes(xlim=(0, 1500), ylim=(0, 1000)) # don't know if these boundaries are correct
    line, = ax.plot([], [], lw=3)

    xcoor = []
    ycoor = []

    for i in data['Coordinates']: 
        xcoor.append(float(i.split(',')[0]))
        ycoor.append(float(i.split(',')[1]))

    data['X']=xcoor
    data['Y']=ycoor

    def init():
        line.set_data([], [])
        return line,

    xdata, ydata = [], [] 

    def animate(i):
        x = data['X'][i]
        y = data['Y'][i]
        xdata.append(x)
        ydata.append(y) 
        line.set_data(xdata, ydata) 
        return line, 

    frame_num = data.shape[0]

    anim = animation.FuncAnimation(fig, animate, init_func=init, frames=frame_num, interval=frame_interval, blit=True)
    anim.save(,riter='imagemagick')
```

```
draw('Neuro.csv', 'Neuro Gif/draw.gif')
```
